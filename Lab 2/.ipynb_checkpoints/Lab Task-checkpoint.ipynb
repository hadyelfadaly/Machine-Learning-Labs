```python
import pandas as pd
import numpy as np
```

```python
def predict(data):

    theta0 = -3.63
    theta1 = 1.16
    
    predictions = [] #numpy matrix we will store our results in

    #iterating over each row in the matrix
    for i in data:

        res = theta0 + theta1*i[0] #applying hypothesis function equation

        predictions.append(float(res))

    return predictions
```

```python
def mean_squared_error(data, predicted_data):

    difference = [] #store sqaured differences
    
    #iterating over the 2 columns simultaneously
    for i,j in zip(data, predicted_data):
        
        difference.append((i-j)**2)

    return np.mean(difference)
```

```python
data = pd.read_csv('data.csv')

#preprocessing
x = data['population'].to_numpy().reshape((-1, 1))
y = data['profit'].to_numpy().reshape((-1, 1))
```

```python
predictions = predict(x)

mse = mean_squared_error(y, predictions)

print(f"Mean Squared Error: {mse}\n")

predictions = pd.DataFrame(predictions, columns=['predictions'])

print('Testing data.csv')
print(f'First 5 Predictions:\n {predictions.head()}')
```
